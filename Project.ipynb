# What are the behaviors of the target audience?

To answer the question about the behaviors of the target audience using the Spotify dataset, we need to consider what aspects of behavior can be inferred from the data. Typically, for music streaming services, audience behavior can be analyzed in terms of their music preferences, like preferred genres, artists, and the popularity of tracks.

```python
#import required libraries
#from google.colab import files
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from scipy.stats import f_oneway
warnings.filterwarnings('ignore')
%matplotlib inline

#Upload and read the csv file
#files.upload()
spotify_df = pd.read_csv('Spotify_data.csv')
```

```python
#Age Group Analysis: Music Genre Preferencs
age_genre_pivot = pd.pivot_table(spotify_df, index='fav_music_genre', columns='Age', aggfunc='size', fill_value=0)
plt.figure(figsize=(14, 8))
sns.heatmap(age_genre_pivot, annot=True, cmap='viridis', fmt='d')
plt.title('Favorite Music Genre accross Different Age Groups')
plt.xlabel('Age Group')
plt.ylabel('Favorite Music Genre')
plt.show()
```

With this heat map we can understand the music and podcast preferences, device usage, and subscription choices across different age groups.

```python
#Gender Analysis: Music Genre Preferences
gender_genre_pivot = pd.pivot_table(spotify_df, index='fav_music_genre', columns='Gender', aggfunc='size', fill_value=0)
plt.figure(figsize=(14, 8))
sns.heatmap(gender_genre_pivot, annot=True, cmap='coolwarm', fmt='d')
plt.title('Favorite Music Genre accross Different Genders')
plt.xlabel('Gender')
plt.ylabel('Favorite Music Genre')
plt.show()
```

In this heatmap we can explore how preferences and behaviors differ across genders.

```python
#Spotify Usage Period Analysis: Music Genre Preferences
usage_genre_pivot = pd.pivot_table(spotify_df, index='fav_music_genre', columns='spotify_usage_period', aggfunc='size', fill_value=0)
plt.figure(figsize=(14, 8))
sns.heatmap(usage_genre_pivot, annot=True, cmap='plasma', fmt='d')
plt.title('Favorite Music Genres accross Different Spotify Usage Periods')
plt.xlabel('Sporify Usage Period')
plt.ylabel('Favorite Music Genre')
plt.show()
```

This heat map gives us an investigation into how long-term users behaviors differ from those of newer users.

These analyses - segmented by age, gender, and Spotify usage period - provide a comprehensive understanding of the behaviors of the target audience. They offer valuable insights into their preferences and habits, which can be crucial for marketing, content creation, and overall user engagement strategies.

# What are the most popular songs? Do they have a common genre, beat, or artist?

```python
#import required libraries
#from google.colab import files
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

#Upload and read the csv file
#files.upload()
spotify_songs_df = pd.read_csv('spotify_songs.csv')
```

```python
#Filter the top 10% most popular songs
top_popular_songs = spotify_songs_df[spotify_songs_df['track_popularity'] >= spotify_songs_df['track_popularity'].quantile(0.90)]
```

```python
#Analyzing common genres for top popular songs
popular_genres = top_popular_songs['playlist_genre'].value_counts()
```

```python
#Analyzing common artists for top popular songs
popular_artists = top_popular_songs['track_artist'].value_counts().head(10)
```

```python
#Analyzing average audio features for top popular songs
avg_audio_features = top_popular_songs[[ 'speechiness', 'acousticness', 'instrumentalness', 'liveness', 'valence']].mean()
```

```python
plt.figure(figsize=(10, 6))
popular_genres.plot(kind='bar', color='skyblue')
plt.title('Common Genres in Top Popular Songs')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Data visualization (Bar Chart) of the common genres for top popular songs

```python
plt.figure(figsize=(10, 6))
popular_artists.plot(kind='bar', color='lightgreen')
plt.title('Top 10 Common Artists in Top Popular Songs')
plt.xlabel('Artist')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

Data visualization (Bar Chart) of the common artist for top popular songs

```python
avg_audio_features.plot(kind='bar', color='salmon')
plt.title('Average Audio Features of Top Popular Songs')
plt.xlabel('Audio Features')
plt.ylabel('Average Value')
plt.xticks(rotation=45)
plt.show()
```

# Is there a best time of year to release a song for maximum popularity?

```python
#import required libraries
#from google.colab import files
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

#Upload and read the csv file
#files.upload()
spotify_songs_df = pd.read_csv('spotify_songs.csv')
```

```python
#Converting release date to datetime
spotify_songs_df['track_album_release_date'] = pd.to_datetime(spotify_songs_df['track_album_release_date'])
```

```python
#Extracting month from the release date
spotify_songs_df['release_month'] = spotify_songs_df['track_album_release_date'].dt.month
```

```python
#Grouping by release month and calculating average popularity
popularity_by_month = spotify_songs_df.groupby('release_month')['track_popularity'].mean()
```

```python
plt.figure(figsize=(12, 6))
popularity_by_month.plot(kind='line', marker='o', color='purple')
plt.title('Average Track Popularity by Release Month')
plt.xlabel('Month')
plt.ylabel('Average Popularity')
plt.xticks(range(1, 13))
plt.show()
```

Data visualization (Line Chart) by month.     
November is the most popular release month.

```python
## going to group up some of the top songs so that it in more balance
# this rounds down to 10 so that we can have it by base 10
def group_track_poularity(num):
    # there is only one song that scores 100 so we are adding that with the 90's
    if num == 100:
        return 90
    return (num - num%10)
```
